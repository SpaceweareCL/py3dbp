```python
from py3dbp import Packer, Bin, Item

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def get_cube():   
    phi = np.arange(1,10,2)*np.pi/4
    Phi, Theta = np.meshgrid(phi, phi)

    x = np.cos(Phi)*np.sin(Theta)
    y = np.sin(Phi)*np.sin(Theta)
    z = np.cos(Theta)/np.sqrt(2)
    return x,y,z

def get_cmap(n, name='hsv'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)



packer = Packer()


packer.add_bin(Bin('large-2-box', 10, 3, 1, 700.0))

packer.add_item(Item('50g [powder 1]', 2, 1, 3, 1))
packer.add_item(Item('50g [powder 1]', 1, 1, 1, 2))
packer.add_item(Item('50g [powder 2]', 1, 1, 1, 3))
packer.add_item(Item('250g [powder 4]', 1, 1, 1, 4))
packer.add_item(Item('250g [powder 5]', 1, 1, 1, 5))
packer.add_item(Item('250g [powder 6]', 1, 1, 1, 6))
packer.add_item(Item('250g [powder 7]', 1, 1, 1, 7))
packer.add_item(Item('250g [powder 8]', 1, 1, 1, 8))
packer.add_item(Item('250g [powder 9]', 1, 1, 1, 9))
packer.add_item(Item('250g [powder 91]', 1, 1, 1, 9))
packer.add_item(Item('250g [powder 92]', 1, 1, 1, 9))
packer.add_item(Item('250g [powder 93]', 1, 1, 1, 9))
packer.add_item(Item('250g [powder 913]', 1, 1, 1, 9))
packer.add_item(Item('250g [powder 923]', 1, 1, 1, 9))
packer.add_item(Item('250g [powder 933]', 1, 1, 1, 9))

packer.pack(bigger_first=True, distribute_items=True)

x,y,z = get_cube()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for ber in packer.bins:
    print(":::::::::::", ber.string())

    print("FITTED ITEMS:")
    cmap = get_cmap(len(ber.items))
    a = float(ber.width)*1.0
    b = float(ber.height)*1.0
    c = float(ber.depth)*1.0  
    
    print(a,b,c)

    d_a = 0
    d_b = 0
    d_c = 0
    ax.plot_surface(x*a+d_a, y*b+d_b, z*c+d_c,  linewidth=0.1, color=cmap(0), alpha=0.3)
    cmpaix = 0
    for item in ber.items:
        
        
        a = float(item.width)*1.0
        b = float(item.height)*1.0
        c = float(item.depth)*1.0  
        
        d_a = float(item.position[0])*1.0
        d_b = float(item.position[1])*1.0
        d_c = float(item.position[2])*1.0
        ax.plot_surface(x*a+d_a, y*b+d_b, z*c+d_c,  linewidth=0.1, color=cmap(cmpaix), alpha=0.3)

        
        print("====> ", (item.position))
        cmpaix =cmpaix+1 


    print("UNFITTED ITEMS:")
    for item in ber.unfitted_items:
        print("====> ", item.position)

    print("***************************************************")
    print("***************************************************")

    
ax.set_xlim(-2,8)
ax.set_ylim(-2,8)
ax.set_zlim(-2,8)
plt.show()
```

```python
from py3dbp import Packer, Bin, Item

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def get_max_depth(bin_used):
    max_width = 0.0
    max_height = 0.0
    max_depth = 0.0
    
    for item in bin_used.items:
        aux_width = float(item.position[0])+float(item.width)
        aux_heigth = float(item.position[1])+float(item.heigth)
        aux_depth = float(item.position[2])+float(item.depth)
        if max_width<aux_width:
            max_width=aux_width
        if max_height<aux_heigth:
            max_height=aux_heigth
        if max_depth<aux_depth:
            max_depth=aux_depth
    
    return max_width, max_height, max_depth


packer = Packer()


packer.add_bin(Bin('large-2-box', 6, 2, 2, 700.0))

packer.add_item(Item('50g [powder 1]', 2, 1, 3, 1))
packer.add_item(Item('50g [powder 2]', 1, 1, 1, 2))
packer.add_item(Item('50g [powder 3]', 1, 1, 1, 3))
packer.add_item(Item('250g [powder 4]', 1, 1, 1, 4))
packer.add_item(Item('250g [powder 5]', 1, 1, 1, 5))
packer.add_item(Item('250g [powder 6]', 1, 1, 1, 6))
packer.add_item(Item('250g [powder 7]', 1, 1, 1, 6))


packer.pack(bigger_first=True, distribute_items=True)

x,y,z = get_cube()
fig = plt.figure()

ax1 = fig.add_subplot(221, projection='3d')

for ber in packer.bins:
    print(":::::::::::", ber.string())

    print("FITTED ITEMS:")
    cmap = get_cmap(len(ber.items))
    
    cmpaix = 0
    for item in ber.items:
        
        
        a = float(item.width)*1.0
        b = float(item.height)*1.0
        c = float(item.depth)*1.0  
        
        d_a = float(item.position[0])*1.0
        d_b = float(item.position[1])*1.0
        d_c = float(item.position[2])*1.0
        ax1.plot_surface(x*a+d_a, y*b+d_b, z*c+d_c,  linewidth=0.1, color=cmap(0), alpha=0.3)

        
        print("====> ", (item.position))
        cmpaix =cmpaix+1 


    print("UNFITTED ITEMS:")
    for item in ber.unfitted_items:
        print("====> ", item.position)

    print("***************************************************")
    print("***************************************************")
    
ax1.set_xlim(-1,6)
ax1.set_ylim(-1,6)
ax1.set_zlim(-1,6)


packer.add_item(Item('250g [powder 78]', 1, 1, 1, 8))
packer.add_item(Item('250g [powder 87]', 1, 1, 1, 8))
packer.add_item(Item('250g [powder 91]', 1, 1, 1, 9))
packer.add_item(Item('250g [powder 92]', 1, 1, 1, 9))
packer.add_item(Item('250g [powder 93]', 1, 1, 1, 9))
packer.add_item(Item('250g [powder 913]', 1, 1, 1, 9))
packer.add_item(Item('250g [powder 923]', 1, 1, 1, 9))
packer.add_item(Item('250g [powder 933]', 1, 1, 1, 9))

ax2 = fig.add_subplot(222, projection='3d')

packer.pack(bigger_first=True, distribute_items=True)

for ber in packer.bins:
    print(":::::::::::", ber.string())

    print("FITTED ITEMS:")
    cmap = get_cmap(len(ber.items))
    
    cmpaix = 0
    for item in ber.items:
        
        
        
        
        a = float(item.width)*1.0
        b = float(item.height)*1.0
        c = float(item.depth)*1.0  
        
        d_a = float(item.position[0])*1.0
        d_b = float(item.position[1])*1.0
        d_c = float(item.position[2])*1.0
        if item.weight<=7:
            ax2.plot_surface(x*a+d_a, y*b+d_b, z*c+d_c,  linewidth=0.1, color=cmap(0), alpha=0.3)
        else:
            ax2.plot_surface(x*a+d_a, y*b+d_b, z*c+d_c,  linewidth=0.1, color=cmap(1), alpha=0.3)

        
        print("====> ", (item.string()))
        cmpaix =cmpaix+1 


    print("UNFITTED ITEMS:")
    for item in ber.unfitted_items:
        print("====> ", item.position)

    print("***************************************************")
    print("***************************************************")

    
ax2.set_xlim(-1,6)
ax2.set_ylim(-1,6)
ax2.set_zlim(-1,6)
    
```
